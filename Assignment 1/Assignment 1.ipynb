```python
#Reading the JSON File
import json
import pandas as pd
with open('ChicagoTraffic.json') as f:
    data = json.load(f)
```

```python
list_json=['row_index', 'unique_identifier', 'zero', 'constant', 'null1', 'constant2', 'null2', 'bracket', 'count', 'direction', 'street', 'constant3', 'constant4', 'East/West', 'latitude', 'longitude', 'list', 'constant5', 'constant6', 'constant7', 'constant8', 'constant9', 'constant10']
```

```python
#Part 2
#Reading the JSON file
data = pd.DataFrame(data=data['data'], columns=list_json)
```

```python
#Q3
Traffic_100_to_115=data.loc[data['street'].isin(['{}th St'.format(f) for f in [100+ f for f in range(16)]])==True]
Traffic_100_to_115['count'] = Traffic_100_to_115['count'].astype(int)
sum(Traffic_100_to_115['count'])
```
